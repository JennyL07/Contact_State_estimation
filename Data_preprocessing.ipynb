```python
import pandas as pd
from datetime import datetime
import numpy as np
import glob
```

```python
df1 = pd.read_csv("imu2.csv")
df1.head()
```

```python
print(df1.iloc[[0, -1], 0], np.shape(df1))
```

```python
df2 = pd.read_csv("joint_states2.csv")
df2.head()
```

```python
print(df2.iloc[[0, -1], 0], np.shape(df2))
```

```python
# drop the first 11 rows of df1 
df2.drop(df2.index[:10], inplace=True)
print(df2.iloc[[0, -1], 0], np.shape(df2))
```

```python
df2.reset_index(drop=True, inplace=True)
print(df2.iloc[[0, -1], 0], np.shape(df2))
```

```python
df2.drop(df2.index[-1], inplace=True)
print(df2.iloc[[0, -1], 0], np.shape(df2))
```

```python
print(df1.iloc[[0, -1], 1])
print(df2.iloc[[0, -1], 1])
```

```python
# Given timestamp
df1["time"] = df1["time"].apply(lambda x: datetime.strptime(x, "%Y/%m/%d/%H:%M:%S.%f").microsecond / 1e3)
df1.head()
```

```python
# check difference between each timestamp
print(df1["time"].diff())
```

```python
# add a column to df1 with difference between each timestamp
df2["time"] = df2["time"].apply(lambda x: datetime.strptime(x, "%Y/%m/%d/%H:%M:%S.%f").microsecond / 1e3)
# df2["diff"] = df2["time"].diff()
df2.head()
```

```python
# check nan values in both df1 and df2
print(df1.isnull().sum(), df2.isnull().sum())
```

```python
# check zeros values in both df1 and df2
print((df1 == 0).sum())
print((df2 == 0).sum())
```

```python
# drop columns 1-9, 13, 17 from df1,
df1.drop(df1.columns[[1, 2, 3, 4, 5, 6, 7, 8, 9, 13, 17]], axis=1, inplace=True)
df1.head()
```

```python
# drop columns 2-6,9 from df2
df2.drop(df2.columns[[1, 2, 3, 4, 5, 8]], axis=1, inplace=True)
df2.head()
```

```python
import ast
# Convert the string representation of tuple to an actual tuple
df2[".position"] = df2[".position"].apply(lambda x: ast.literal_eval(x))

# Expand tuple into separate columns
df2_expanded = df2[".position"].apply(pd.Series)

# Rename columns for clarity
df2_expanded.columns = [f"position_{i}" for i in range(df2_expanded.shape[1])]

# Ensure all values are float
df2_expanded = df2_expanded.astype(float)

# Combine with original DataFrame if needed
df2 = pd.concat([df2, df2_expanded], axis=1).drop(columns=[".position"])
```

```python
df2[".velocity"] = df2[".velocity"].apply(lambda x: ast.literal_eval(x))

# Expand tuple into separate columns
df2_expanded = df2[".velocity"].apply(pd.Series)

# Rename columns for clarity
df2_expanded.columns = [f"velocity_{i}" for i in range(df2_expanded.shape[1])]

# Ensure all values are float
df2_expanded = df2_expanded.astype(float)

# Combine with original DataFrame if needed
df2 = pd.concat([df2, df2_expanded], axis=1).drop(columns=[".velocity"])

df2.head()
```

```python
# contact two dataframes
df1.drop(df1.columns[[0]], axis=1, inplace=True)
df2.drop(df2.columns[[0]], axis=1, inplace=True)
df = pd.concat([df1, df2], axis=1)
df.head()
```

```python
# read csv file name as df3
df3 = pd.read_csv("converted_contact2.csv", delimiter="\t", header=None, names=["left", "right"])
df3.head()
```

```python
# print shape of df3
print(np.shape(df3))
# check element type in df3
print(df3.dtypes)
print(df3.iloc[[0, -1], 0], np.shape(df3))
```

```python
df3.drop(df3.index[:11], inplace=True)
df3.reset_index(drop=True, inplace=True)
print(np.shape(df3))
```

```python
# contact df3 to df
df = pd.concat([df, df3], axis=1)
df.head()
```
